THE DATASET IS UNBALANCED AS THE OUTPUT MAJORITY IS HIGHLY ON ONE SIDE
0 --> NON-FRAUD TRANSACTION
1 --> FRAUD TRANSACTION 

```python
#seperating the data for analysis
legit = cc_data[cc_data.Class == 0]
fraud = cc_data[cc_data.Class == 1]
```

UNDERSAMPLING WILL BE USED WHEN WE WANT THE MAJORITY OUPUT BE 1:1 WITH THE MINORITY OUPUT (n=no.of minority data) 
OVERSAMPLING WILL BE USED WHEN WE WANT TO INCREASE THE  MINORITY OF THE DATA 1:1 WITH THE MAJORITY OUTPUT (USING SMOT)

EXAMPLE OF USING OVERSAMPLING (SMOTE)

```python
%pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
#using oversampling to balance the data to majority class
X_res, Y_res = SMOTE().fit_resample(cc_data.drop('Class', axis=1), cc_data['Class'])
Y_res.value_counts()
```

THE MODEL IS TRAINED USING THE UNDER SAMPLED DATA 

```python
#using undersampling to balance the data
#random sampling
legit_sample = legit.sample(n=492)
```

```python
#concatenating two dataframes
new_cc_data = pd.concat([legit_sample, fraud], axis=0)
```

```python
#checking the new data
new_cc_data.head()
```

```python
new_cc_data.tail()
```

```python
#new data value counts
new_cc_data['Class'].value_counts()
```

CHECKING THE MEAN OF BOTH THE DATASET TO CHECK IF THE NATURE OF THE DATASET REMAINS THE SAME 

```python
#OLD data mean 
cc_data.groupby('Class').mean()
```

```python
#new data mean
new_cc_data.groupby('Class').mean()
```

```python
#splitting the data into features and target
X = new_cc_data.drop(columns='Class', axis=1)
Y = new_cc_data['Class']
```

```python
print (X)
```

```python
print (Y)
```

```python
#splitting the data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print (X.shape, X_train.shape, X_test.shape)
```

```python
# Train the model using logistic regression
model = LogisticRegression(max_iter=500)
model.fit(X_train, Y_train)
```

```python
#accuracy on training data
model.score(X_train, Y_train)
```

```python
#accuracy on test data
model.score(X_test, Y_test)
```

```python
#training the model using the decision tree
from sklearn.tree import DecisionTreeClassifier
```

```python
model2 = DecisionTreeClassifier()
model2.fit(X_train, Y_train)
```

```python
#accuracy on training data
model2.score(X_res, Y_res)
```

```python
#accuracy on test data
model2.score(X_test, Y_test)
```

```python
#As the Train accuracy and test accuracy is close to each other in both the models, we can say that the model is not overfitting.
#Now we will check the confusion matrix and classification report for both the models

confusion_matrix(Y_test, model.predict(X_test))
print (classification_report(Y_test, model.predict(X_test)))
confusion_matrix(Y_test, model2.predict(X_test))
print (classification_report(Y_test, model2.predict(X_test)))
```

```python
#the model (logistic regression) is giving better results than the decision tree model, as the accuracy is higher.
```
